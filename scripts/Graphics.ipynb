```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = pd.read_csv("data/train.csv")
data["PRI_jet_num"]
```

```python
res_0 = np.genfromtxt(
    "data/case_0_CV5_statistics_median_null_var_pairwise_bias_scaling_orth_pol_1_to_20_1e4.csv", delimiter=",", skip_header=1)
mean_0 = np.array([res_0[i][0] for i in range(res_0.shape[0])])
median_0 = np.array([res_0[i][1] for i in range(res_0.shape[0])])
std_0 = np.array([res_0[i][2] for i in range(res_0.shape[0])])
error = 100*1.96*std_0
fig, ax0 = plt.subplots(nrows=1, sharex=True)
ax0.errorbar(np.arange(1, res_0.shape[0]+1), mean_0*100, yerr=error, fmt='-o')
ax0.set_title('PRI_jet_num = 0 ~ Polynomials from 1 to 20')
plt.xticks(np.arange(1, res_0.shape[0]+1))
plt.hlines(y = max(mean_0*100), xmin = 1, xmax = np.argmax(mean_0*100)+1, ls=':')
plt.vlines(x=np.argmax(mean_0*100)+1, ymin = 82, ymax = max(mean_0*100), ls=':')
plt.show()
```

```python
index = np.argmax(mean_0)
print(mean_0[index]*100, std_0[index]*100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x+y

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```
